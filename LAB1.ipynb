```python
import numpy as np
import math 
import matplotlib.pyplot as plt

C = 0.001
R = 1802

alfa = 1 / (C * R)
print("Alfa = ", alfa)

t = np.log(20)*alfa
print("Czas po ktorym odpowiedz swobodna ukladu nie przekracza 0.05 = ", t)

#############################################################################

C = 0
alfa = 1

def funkcja(omega):
  print("Wykres dla = ", omega)
  t = np.linspace(0.1, 10, 100)

  corj = C*np.exp(-alfa * t)
  corjp = -C * alfa * np.exp(alfa * t)

  x = corj + (alfa * np.sin(omega * t) - omega * np.cos(omega * t))/ (omega * omega + alfa * alfa)
  xp = corjp + (alfa * omega * np.cos(omega * t) + omega * omega * np.sin(omega * t)) / (omega * omega + alfa * alfa)

  u = np.sin(omega * t)
  fig, ax = plt.subplots()
  ax.plot(t, x, t, xp, t, u)

  ax.grid(True)
  plt.xscale('log') 
  plt.show()

funkcja(0.1)
funkcja(1)
funkcja(3)
funkcja(5)
funkcja(7)
funkcja(10)  
```
